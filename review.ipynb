导航数据说明：
每个骑士在每一个波次的导航数据包括
1 骑士在被assign订单是所在的位置点
2 每个订单的商家的位置
3 每个订单的客户的位置
例如在一个波次中一个骑士接受了两个单子，那么该骑士接受订单的位置点就会有两个，商户位置有2个，客户位置有2个，共计有6个点
那么这6个点的导航线路共有A62=6*5=30个
